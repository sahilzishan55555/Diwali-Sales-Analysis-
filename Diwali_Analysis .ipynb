# EXPLORATORY DATA ANALYSIS PART

```python
dataframe.columns
```

ON THE BASIS OF "GENDER"

```python
ax = sns.countplot(x="Gender", data=dataframe)   # Code is used to plot a count plot or a bar graph for column gender 
                                # Using this plot we can see that female customers are more as compared to the male customers 
for bars in ax.containers:
    ax.bar_label(bars)                              

plt.show()
```

SALES TOTAL AS PER GENDER 

```python
sales_gen = dataframe.groupby(['Gender'], as_index=False)['Sales_Amount'].sum().sort_values(by='Sales_Amount', ascending=False)

sns.barplot(x = 'Gender',y= 'Sales_Amount' ,data = sales_gen)

# We can conclude for the graph that female customers are buying more and purchasing power of female is more
# Total sales amount in female is more
```

ONE THE BASIS OF AGE

```python
ax = sns.countplot(data = dataframe, x = 'Age Group', hue = 'Gender')

for bars in ax.containers:
    ax.bar_label(bars)
    
    # from the below plot we can conclude that female customers are purchasing more in age group of 26 to 35
```

ON THE BASIS OF TOTAL AMOUNT AND AGE GROUP 

```python
sales_age = dataframe.groupby(['Age Group'], as_index=False)['Sales_Amount'].sum().sort_values(by='Sales_Amount', ascending=False)

sns.barplot(x = 'Age Group',y= 'Sales_Amount' ,data = sales_age)

# Analyzing below graph we can see that age group of 26 to 35 are purchasing more and contributing more in sales and revnue
```

SALES ON THE BASIS OF DIFFERENT STATES

```python
sales_state = dataframe.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data = sales_state, x = 'State',y= 'Orders')

# Top 3 states are utter pradesh,maharashtra, karnataka who are giving top revnue 
```

SALES AND REVNUE ON THE BASIS OF MARITAL STATUS 

```python
ax = sns.countplot(data = dataframe, x = 'Marital_Status')

sns.set(rc={'figure.figsize':(7,5)})
for bars in ax.containers:
    ax.bar_label(bars)
    
    # Unmarried people are shopping more and more unmarried people are the customers as compared to married once
```

```python
sales_state = dataframe.groupby(['Marital_Status', 'Gender'], as_index=False)['Sales_Amount'].sum().sort_values(by='Sales_Amount', ascending=False)

sns.set(rc={'figure.figsize':(6,5)})
sns.barplot(data = sales_state, x = 'Marital_Status',y= 'Sales_Amount', hue='Gender')
```

ON THE BASIS OF OCCUPATIONS

```python
sns.set(rc={'figure.figsize':(20,5)})
ax = sns.countplot(data = dataframe, x = 'Occupation')

for bars in ax.containers:
    ax.bar_label(bars)
    
    # Most customers are from IT sector, healthcae and avaition
```

ON THE BASIS OF PRODUCT CATEGORY 

```python
sns.set(rc={'figure.figsize':(20,5)})
ax = sns.countplot(data = dataframe, x = 'Product_Category')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state = dataframe.groupby(['Product_Category'], as_index=False)['Sales_Amount'].sum().sort_values(by='Sales_Amount', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = 'Product_Category',y= 'Sales_Amount')

#Top 3 product category are Food, clothing and electronics
```
